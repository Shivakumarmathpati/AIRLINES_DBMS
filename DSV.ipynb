```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.rand(1000)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 12))

# Plot the histogram
ax1.hist(data, bins=30, alpha=0.5, label='Histogram')
ax1.set_title('Histogram')
ax1.legend()

# Plot the box plot
ax2.boxplot(data)
ax2.set_title('Box plot')

# Display the plots
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import Legend

output_notebook()

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a new plot
p = figure(title="Bokeh Line Graph with Annotations and Legends")

# Add a line renderer with legend and line thickness
line = p.line(x, y, legend_label="Line", line_width=2)

# Customize the legend location
legend = Legend(items=[("Line", [line])], location="center")
p.add_layout(legend, 'right')

# Show the results
show(p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = np.array(['A', 'B', 'C', 'D', 'E'])
values = [4, 3, 2, 5, 4]
values = np.append(values, values[0])

# Calculate angle for each axis
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.fill(angles, values, color='red', alpha=0.25)
ax.plot(angles, values, color='red', linewidth=2)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Radar chart')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Example data
data = {
    'state': ['CA', 'TX', 'NY', 'FL', 'IL'],
    'value': [10, 20, 30, 40, 50]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a choropleth map
fig = px.choropleth(df,
                    locations='state',
                    locationmode='USA-states',
                    color='value',
                    scope='usa',
                    title='Sample Choropleth Map of the USA',
                    color_continuous_scale='Blues')

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a pair plot (correlogram)
sns.pairplot(tips)
plt.suptitle('Correlogram', y=1.02)  # Adding title above the pair plot
plt.show()

# Calculate the correlation matrix
corr = tips.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
